```python
# imports
import os
import glob
import numpy as np
import pandas as pd 
import altair as alt
# import altair_viewer
```

```python
# load in dataframes and concatenate together
path = (
    "C:/Users/jacob.hester/Documents/Python_scripts/"\
    "Historical_beef_cattle_prices/data/processed"
    )
raw_files_path = os.path.join(path, "*.csv")
df_list = []
for f in glob.glob(raw_files_path):
    # print(f)
    df = pd.read_csv(f, index_col=None, header=0)
    df_list.append(df)
df = pd.concat(df_list, ignore_index = True)
```

```python
df.head() 
```

```python
# create Overall_type colum
df[['Overall_type', 'Sub_type']] = df['Type'].str.split(':', 1, expand=True)
print(df.head())
```

```python
feeder_df = df[df['Overall_type'].isin(['Feeder Bulls', 'Feeder Heifers', 'Feeder Steers'])]
```

```python
# select columns to keep and drop missing values
cols_to_keep = ['Year', 'Overall_type', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
feeder_df = feeder_df[cols_to_keep]
feeder_df = feeder_df.dropna() 
```

```python
feeder_df.head() 
```

```python
# reshape from wide to long
feeder_df = pd.melt(
    feeder_df, 
    id_vars = ['Year', 'Overall_type'], 
    value_vars = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)
print(feeder_df.head(50))
```

```python
# get average value by group for each year, month
feeder_avg_df = feeder_df.groupby(['Year', 'Overall_type', 'variable'])['value'].mean()
```

```python
feeder_avg_df = pd.DataFrame(feeder_avg_df)
feeder_avg_df.reset_index(inplace = True)
```

```python
feeder_avg_df.head(50)
```

```python
# date cols
feeder_avg_df['Year'] = pd.to_datetime(
    feeder_avg_df['Year'].astype(str), format='%Y'
)
# add column for month and year 
months_dict = {'Jan' : 1, 
               'Feb' : 2,
               'Mar' : 3, 
               'Apr' : 4, 
               'May' : 5, 
               'Jun' : 6, 
               'Jul' : 7, 
               'Aug' : 8, 
               'Sep' : 9, 
               'Oct' : 10, 
               'Nov' : 11, 
               'Dec' : 12
              }
feeder_avg_df['month'] = feeder_avg_df['variable'].map(months_dict)
feeder_avg_df['year'] = feeder_avg_df['Year'].dt.year
feeder_avg_df['day'] = 1
feeder_avg_df['date_col'] = pd.to_datetime(feeder_avg_df[['month', 'year', 'day']])
```

```python
print(feeder_avg_df.head(10))
print('')
print(feeder_avg_df.dtypes)
```

```python
# create year, overall_type column
change_data_types = {'year' : str}
feeder_avg_df = feeder_avg_df.astype(change_data_types)
feeder_avg_df['year_type'] = feeder_avg_df['year'] + feeder_avg_df['Overall_type']
feeder_avg_df.head()
```

```python
# look at what month is highest each year 
max_vals_dict = feeder_avg_df.groupby(['year_type'])['value'].max().to_dict()
print(max_vals_dict)
```

```python
# create max price column
feeder_avg_df['max_price'] = feeder_avg_df['year_type'].map(max_vals_dict)
feeder_avg_df['max_price_month'] = np.where(feeder_avg_df['value'] == feeder_avg_df['max_price'], 1, 0)
feeder_avg_df.head(20)
```

```python
# visual
feeder_avg_per_month = alt.Chart(feeder_avg_df).mark_point().encode(
     color = 'max_price_month',
    x = alt.X(
        'date_col', 
        axis = alt.Axis(
            title = 'Date',
            titleFontSize = 14
            ),
            ),
    y = alt.Y(
        'value', 
        axis = alt.Axis(
            title = 'Average Price',
            titleFontSize = 14
            )
        )
    ).properties(
            title = {
                "text" : "Average Feeder Price Per Month",
                "subtitle" : "Average Monthly Beef Cattle Price in Alabama, 2004-2018 by Type"
                },
            width = 2000
    )

feeder_text = (
    alt.Chart(feeder_avg_df.query("max_price_month == 1"))
    .mark_text(dy=-15, color="darkblue")
    .encode(x=alt.X("date_col"), y=alt.Y("value"), text=alt.Text("variable"))
)

(feeder_avg_per_month + feeder_text)
```
